```python
import requests
from bs4 import BeautifulSoup

def phoneTel(url):
    try:
       r = requests.get(url)
       soup = BeautifulSoup(r.content, 'html.parser')
       x = soup.find_all('p', {'itemprop' : 'telephone'})
       for i in x:
           y = i.text.replace('\n', '')
           return y
    except:
          pass
          y = ('Nan')
          return  y

print(phoneTel("https://www.usine-digitale.fr/annuaire-start-up/izi-family,1072689"))
```
